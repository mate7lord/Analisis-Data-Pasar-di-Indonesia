```python
# Impor Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
# Memuat Data
file_path = 'Data Pasar di Indonesia 2024-08-02.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')
```

```python
# Eksplorasi Data
print("Dimensi Data:", data.shape)
print("Info Data:")
print(data.info())
print("Beberapa baris pertama dari data:")
print(data.head())
```

```python
# Pembersihan Data
# Memeriksa nilai yang hilang
print("Nilai yang Hilang:")
print(data.isnull().sum())
```

```python
# Mengisi atau Menghapus nilai yang hilang (contoh: mengisi 'legalitas_pasar' yang hilang dengan 0)
data['legalitas_pasar'].fillna(0, inplace=True)
```

```python
# Statistik Deskriptif
print("Statistik Deskriptif:")
print(data.describe())
```

```python
# Visualisasi Data
# Jumlah pasar per provinsi
plt.figure(figsize=(14, 7))
sns.countplot(y='provinsi', data=data, order=data['provinsi'].value_counts().index)
plt.title('Jumlah Pasar per Provinsi')
plt.xlabel('Jumlah')
plt.ylabel('Provinsi')
plt.show()
```

```python
# Analisis Sertifikasi Pasar
pasar_tersertifikasi = data[data['legalitas_pasar'] == 1]
pasar_tidak_tersertifikasi = data[data['legalitas_pasar'] == 0]

print("Pasar Tersertifikasi:", pasar_tersertifikasi.shape[0])
print("Pasar Tidak Tersertifikasi:", pasar_tidak_tersertifikasi.shape[0])
```

```python
# Visualisasi sertifikasi pasar
plt.figure(figsize=(7, 7))
sns.countplot(x='legalitas_pasar', data=data)
plt.title('Status Sertifikasi Pasar')
plt.xlabel('Legalitas Pasar')
plt.ylabel('Jumlah')
plt.xticks(ticks=[0, 1], labels=['Tidak Tersertifikasi', 'Tersertifikasi'])
plt.show()
```

# Kesimpulan dan Wawasan

Analisis data pasar di Indonesia mengungkapkan beberapa wawasan kunci:
1. Distribusi Pasar: Jumlah pasar bervariasi secara signifikan di berbagai provinsi, dengan beberapa provinsi memiliki konsentrasi pasar yang lebih tinggi.
2. Status Legal: Sebagian besar pasar tidak memiliki sertifikasi legal, menunjukkan potensi area untuk perbaikan regulasi.
3. Karakteristik Pasar: Statistik deskriptif memberikan gambaran umum tentang karakteristik pasar, seperti jumlah pasar per kota dan status operasionalnya.

Wawasan ini dapat membantu pembuat kebijakan dan bisnis dalam membuat keputusan yang tepat terkait pengembangan dan regulasi pasar di Indonesia.
